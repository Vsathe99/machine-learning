```python

def check_number(num):
    if num > 0:
        print(f"{num} is a positive number.")
    elif num == 0:
        print(f"{num} is zero.")
    else:
        print(f"{num} is a negative number.")


def print_numbers_for_loop(limit):
    print("Printing numbers using a for loop:")
    for i in range(limit):
        print(i, end=" ")



def print_numbers_while_loop(limit):
    print("Printing numbers using a while loop:")
    i = 0
    while i < limit:
        print(i, end = " ")
        i += 1


def loop_with_break(limit):
    print("Loop with break statement:")
    for i in range(limit):
        if i == 5:
            print("Breaking out of the loop at i =", i)
            break
        print(i, end=" ")


check_number(10)
check_number(-5)
check_number(0)

print_numbers_for_loop(5)
print()
print_numbers_while_loop(5)
print()
loop_with_break(10)
```

```python
import pandas as pd
import numpy as np

# 1. Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Salary': [70000, 80000, np.nan, 60000, 75000]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n", df)


print("\nDataFrame Info:")
print(df.info())


print("\nDescriptive Statistics (Summary):")
print(df.describe())


df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print("\nDataFrame after filling missing values:\n", df)


ages = df['Age']
print("\nAges Column:\n", ages)


filtered_df = df[df['Age'] > 25]
print("\nFiltered DataFrame (Age > 25):\n", filtered_df)


sorted_df = df.sort_values(by='Age')
print("\nDataFrame Sorted by Age:\n", sorted_df)


grouped_df = df.groupby('City')['Salary'].mean()
print("\nAverage Salary Grouped by City:\n", grouped_df)


df['Age Group'] = ['Young' if x < 30 else 'Mature' for x in df['Age']]
print("\nDataFrame with New 'Age Group' Column:\n", df)


df['Salary'] = df['Salary'].apply(lambda x: x * 1.1)
print("\nDataFrame after applying 10% salary raise:\n", df)


df = df.drop(columns=['Age Group'])
print("\nDataFrame after Dropping 'Age Group' Column:\n", df)


data2 = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Department': ['HR', 'Finance', 'Engineering', 'Marketing', 'IT']
}

df2 = pd.DataFrame(data2)

merged_df = pd.merge(df, df2, on='Name', how='left')
print("\nMerged DataFrame with Department Information:\n", merged_df)


pivot_table = pd.pivot_table(df, values='Salary', index='City', aggfunc=np.mean)
print("\nPivot Table - Average Salary by City:\n", pivot_table)


df.rename(columns={'Salary': 'Annual Salary'}, inplace=True)
print("\nDataFrame with Renamed Columns:\n", df)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline


# Sample dataset
data = {
    'Age': [25, 30, np.nan, 35, 40],
    'Salary': [50000, 60000, 70000, np.nan, 90000],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female']
}

df = pd.DataFrame(data)

# Impute missing values in numerical columns with mean
imputer = SimpleImputer(strategy='mean')
df[['Age', 'Salary']] = imputer.fit_transform(df[['Age', 'Salary']])

print(df)

# Scaling numerical features using StandardScaler
scaler = StandardScaler()
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])

print(df)

# One-Hot Encoding for categorical column
encoder = OneHotEncoder(drop='first')  # drop first to avoid multicollinearity
encoded_gender = encoder.fit_transform(df[['Gender']])

# Convert to DataFrame
encoded_gender_df = pd.DataFrame(encoded_gender, columns=encoder.get_feature_names_out(['Gender']))
df = df.join(encoded_gender_df).drop('Gender', axis=1)

print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Center the data
def center_data(X):
    return X - np.mean(X, axis=0)

# Step 2: Compute the covariance matrix
def covariance_matrix(X_centered):
    return np.cov(X_centered, rowvar=False)

# Step 3: Compute eigenvalues and eigenvectors
def compute_eigen(cov_matrix):
    eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
    return eigenvalues, eigenvectors

# Step 4: Sort eigenvectors by eigenvalues
def sort_eigenvectors(eigenvalues, eigenvectors):
    sorted_indices = np.argsort(eigenvalues)[::-1]
    sorted_eigenvalues = eigenvalues[sorted_indices]
    sorted_eigenvectors = eigenvectors[:, sorted_indices]
    return sorted_eigenvalues, sorted_eigenvectors

# Step 5: Project data onto principal components
def project_data(X_centered, eigenvectors, k):
    return np.dot(X_centered, eigenvectors[:, :k])

# PCA Class Implementation
class PCA:
    def __init__(self, n_components):
        self.n_components = n_components
        self.mean_ = None
        self.components_ = None
        self.explained_variance_ = None

    def fit(self, X):
        # Step 1: Center the data
        X_centered = center_data(X)

        # Step 2: Compute the covariance matrix
        cov_matrix = covariance_matrix(X_centered)

        # Step 3: Compute eigenvalues and eigenvectors
        eigenvalues, eigenvectors = compute_eigen(cov_matrix)

        # Step 4: Sort eigenvectors by eigenvalues
        sorted_eigenvalues, sorted_eigenvectors = sort_eigenvectors(eigenvalues, eigenvectors)

        # Step 5: Save the results
        self.mean_ = np.mean(X, axis=0)
        self.components_ = sorted_eigenvectors[:, :self.n_components]
        self.explained_variance_ = sorted_eigenvalues[:self.n_components]

    def transform(self, X):
        # Center the data
        X_centered = center_data(X)

        # Project the data onto the principal components
        return project_data(X_centered, self.components_, self.n_components)

    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)

# Example usage

# Create a sample dataset (5 features and 10 samples)
np.random.seed(42)
X = np.random.rand(10, 5)  # 10 samples, 5 features

# Perform PCA to reduce the dimensionality to 2
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Show the results
print("Original data shape:", X.shape)
print("Transformed data shape:", X_pca.shape)

# Plotting the original data
plt.figure(figsize=(12, 6))

# Plot original data in 5D (this is just a 2D projection for illustration)
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], color='b', label='Original Data')
plt.title('Original Data (Projection of First Two Features)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plotting the transformed data (PCA results) in 2D
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], color='r', label='PCA Transformed Data')
plt.title('PCA Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.tight_layout()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import make_classification

# Step 1: Create or load a dataset
# We'll generate a random dataset for illustration
np.random.seed(42)
X, _ = make_classification(n_samples=100, n_features=5, random_state=42)

# Step 2: Standardize the data (important for PCA)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply PCA
pca = PCA(n_components=2)  # Reduce the data to 2 dimensions
X_pca = pca.fit_transform(X_scaled)

# Step 4: Visualize the results
plt.figure(figsize=(8, 6))

# Plot the original data in the 2D space after PCA transformation
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='r', marker='o')
plt.title('PCA Transformed Data (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()

# Display explained variance ratio to understand how much variance is explained by each principal component
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Cumulative Explained Variance:", np.cumsum(pca.explained_variance_ratio_))

```
