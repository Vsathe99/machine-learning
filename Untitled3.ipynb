```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
x = np.array([29,15,33,28,39]).reshape(-1,1)
y = np.array([0,0,1,1,1])
```

```python
model = LogisticRegression()
model.fit(x,y)
```

```python
hours = np.array([10]).reshape(-1,1)
probability = model.predict_proba(hours)[0][1]
decision = model.predict(hours)[0]

print(f"Probability: {probability:.3f}")
print(f"Decision: {'yes' if decision == 1 else 'no'}")
```

```python
plt.scatter(x[:,0],y, color = 'red', label = 'Data Points')
plt.xlabel("No of study hours")
plt.ylabel("result (1=yes, 0=no)")
plt.title("logestic regression")

hours = np.linspace(10,50,100).reshape(-1,1)
x_test = hours
probabilites = model.predict_proba(x_test)[:,1]

plt.plot(hours,probabilites, color = 'blue', label = 'probability curve')
plt.legend()
plt.show()
```
